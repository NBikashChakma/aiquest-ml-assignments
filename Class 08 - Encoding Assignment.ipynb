# Loop concept for LabelEncoder

```python
df2.head()
```

```python
#Type 1 Loop

for col in df2.columns:
    if df2[col].dtype == np.number:
        continue
    else:
        df2[col] = label_encoder.fit_transform(df[col])
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for col in df6.columns:
    if is_numeric_dtype(df6[col]): #is_numeric_dtype is pandas keyword which check is this is number or not 
        continue
    else:
        one = pd.get_dummies(df6[col], drop_first=True,prefix='Area')
        df6.drop(df6[[col]], axis=1, inplace=True)
        df6 = pd.concat([df6, one], axis=1)
```

# Loop

```python
for col in df8.columns:
    if is_numeric_dtype(df8[col]):
        continue
    else:
        order=df8[col].unique()
        ordinal = OrdinalEncoder(categories=[order])
        encoded = ordinal.fit_transform(df8[[col]])
        df8[col] = encoded
```
